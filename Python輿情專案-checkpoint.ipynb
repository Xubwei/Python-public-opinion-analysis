# 利用K-means分群

```python
from sklearn.cluster import KMeans
```

```python
# 將「評價_推」、「評價_中立」與「評價_噓」的次數加總起來，即為該貼文的「聲量」
underwearData['聲量'] = underwearData.filter(regex = '評價').sum(axis =1)
```

```python
# 制定每一筆資料的「資料流水號」，便於後續統計每一筆貼文下留言的相關聲量
underwearData['資料流水號'] = range(1, len(underwearData)+1)
```

### 以留言為基底去做分析

```python
# 拆解貼文的留言，並變成以留言為基底的資料模式，以便後續統計消費者的顧客樣貌
newlist = []
for i in underwearData['資料流水號']:
    underwearDatatmp = underwearData[underwearData['資料流水號']==i]
    try:
        new = pd.DataFrame(eval(underwearDatatmp['留言'].iloc[0]))
        new['標題'] = underwearDatatmp['標題'].iloc[0]
        new['內文'] =  underwearDatatmp['內文'].iloc[0]
        new['聲量'] =  underwearDatatmp['聲量'].iloc[0]
        newlist.append(new)
    except Exception  as e:
        print(e)
```

```python
# 將留言合併成dataframe
underwearData = pd.concat(newlist)
# 將貼文與留言合併到一個欄位
underwearData['所有文'] = underwearData['標題'] + underwearData['內文']+underwearData['content']
```

```python
underwearData
```

```python
# 要以【消費者為中心考量】
# 產出每一個Tag欄位，計算每一個商品被該Tag提到與否
```

```python
# 在underwearData中新增一系列feature詞頻
mentioned_features = []
for i in feature:
    # 如果有含Tag，就算１
    underwearData[i] = np.where(underwearData['所有文'].str.contains(i, na=False), 1, 0)

    # 記錄我們自訂tag裏面有多少個符合消費者標籤
    # 原因：有些字詞是合作方自提，但實際上，消費資料中的tag沒有
    if len(underwearData[underwearData['所有文'].str.contains(i, na=False)])>0:
        mentioned_features.append(i)
```

```python
# 查看【自訂tag】與符合消費者標籤的個數
print('總共有' +str(len(feature))+'個自訂tag，' +'但實際上只有'+  str(len(mentioned_features))+'個tag符合消費者標籤的個數' )
```

## K-means 前處理

```python
# 將要計算的【消費價錢】（買方_價格）與特徵字詞加入同一個list裏面
all_features = ['聲量']
```

```python
#將all_features與mentioned_features的list融合在一起
# 變成我們想要的【特徵字詞】
all_features.extend(mentioned_features)
```

```python
# 將每一個使用者提到的關鍵字加總，變成使用者關鍵字表
underwear_data_cluster= underwearData.groupby(['user'], as_index=False)[all_features].sum()
```

```python
underwear_data_cluster
```

```python
# 將'買家ID','聲量'設定爲index，這樣做分群的時候就不會干擾到
underwear_data_cluster = underwear_data_cluster.set_index(['user', '聲量'])
```

## K-means

### 機器學習自動決定要分多少群

```python
# 機器學習自動決定要分多少群
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k,random_state=1)
    kmeanModel.fit(underwear_data_cluster)
    distortions.append(kmeanModel.inertia_)
```

```python
# 視覺化
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
distortions
```

## 決定選取8群

```python
clustering = KMeans(n_clusters=8,random_state=1)
clustering.fit(underwear_data_cluster)
```

```python
# 查看分群標籤
clustering.labels_
```

```python
# 將群放入underwear_data_cluster
underwear_data_cluster['群'] = clustering.labels_
```

```python
underwear_data_cluster
```
